```python
result7 = pd.merge(result6, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_3', right_on='provider_id', suffixes=['_2', '_3'])
```

```python
result8 = pd.merge(result7, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_4', right_on='provider_id', suffixes=['_3', '_4'])
```

```python
result9 = pd.merge(result8, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_5', right_on='provider_id', suffixes=['_4', '_5'])
```

```python
result9['row_rating'].fillna(0,inplace=True)
result9['row_rating_1'].fillna(0,inplace=True)
result9['row_rating_2'].fillna(0,inplace=True)
result9['row_rating_3'].fillna(0,inplace=True)
result9['row_rating_4'].fillna(0,inplace=True)
result9['years_experience'].fillna(0,inplace=True)
result9['prov_qual_indic'].fillna(0,inplace=True)
result9['quality_measures'].fillna('N',inplace=True)
result9['electronic_measures'].fillna('N',inplace=True)
result9.ix[result9['quality_measures'] =='0','quality_measures']='N'
result9.ix[result9['electronic_measures'] =='0','electronic_measures']='N'
#result9.fillna(0, inplace=True)
result9['hmaxscore'] = result9[['row_rating','row_rating_1','row_rating_2','row_rating_3','row_rating_4']].apply(max, axis=1)
```
