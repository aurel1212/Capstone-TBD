```python
result5 = pd.merge(result4, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_1', right_on='provider_id', suffixes=['_lll', '_1'])
```

```python
result6 = pd.merge(result5, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_2', right_on='provider_id', suffixes=['_1', '_2'])
```

```python
result7 = pd.merge(result6, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_3', right_on='provider_id', suffixes=['_2', '_3'])
```

```python
result8 = pd.merge(result7, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_4', right_on='provider_id', suffixes=['_3', '_4'])
```

```python
result9 = pd.merge(result8, wa_hosp_rating_gr, how='left',left_on='related_org_ccn_5', right_on='provider_id', suffixes=['_4', '_5'])
```

```python
#final matrix
result9.head(10).T
```

```python
keep_columns=['npi_l', 'last_org_name', 'first_name_l', 'mi', 
 'credentials', 'gender_l',  'state_l', 'type', 'report_year', 'zip_base', 
 'bene_day_srvc_cnt', 'submitted_delta_perc','allowed_delta_perc', 
 'pac_id_ll', 'credential', 'medical_school',  'primary_specialty', 
 'quality_measures', 'electronic_measures', 'years_experience', 'prov_qual_indic', 
 'provider_id_1','hospital_name_1','row_rating_1', 
 'provider_id_2', 'hospital_name_2', 'row_rating_2',
 'provider_id_3', 'hospital_name_3', 'row_rating_3',
 'provider_id_4', 'hospital_name_4', 'row_rating_4',
 'provider_id','hospital_name', 'row_rating']
```
